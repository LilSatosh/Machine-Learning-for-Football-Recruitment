```python
#import standard stats table for the Top5 European 
#football for 2024/25 season
import pandas as pd
import numpy as np
dataset_standard = pd.read_csv("fbref_big5_stats_24_25_standard.csv")
print(dataset_standard)
```

```python
#rename columns in standard stats table to easy to undertand words using Fbref glossary 
rename_columns = {
    "Rk": "Rank",
    "Nation": "Nationality",
    "Pos": "Position",
    "Comp": "Competition",
    "Age": "Age",
    "Born": "Birth_Year",
    
    #playing time
    "MP": "Matches_Played",
    "Starts": "Games_Started",
    "Min": "Minutes",
    "90s": "90s_Played",
    
    #performance
    "Gls": "Goals",
    "Ast": "Assists",
    "G+A": "Goals_Plus_Assists",
    "G-PK": "Non_Penalty_Goals",
    "PK": "Penalties_Made",
    "PKatt": "Penalties_Attempted",
    "CrdY": "Yellow_Cards",
    "CrdR": "Red_Cards",
    
    #expected metrics
    "xG": "Expected_Goals",
    "npxG": "Non_Penalty_xG",
    "xAG": "Exp_Assisted_Goals",
    "npxG+xAG": "Non_Penalty_xG_Plus_xAG",
    
    #progression
    "PrgC": "Progressive_Carries",
    "PrgP": "Progressive_Passes",
    "PrgR": "Progressive_Passes_Received",

    #stats PER 90
    "Gls.1": "Goals_Per_90",
    "Ast.1": "Assists_Per_90",
    "G+A.1": "Goals_Plus_Assists_Per_90",
    "G-PK.1": "Non_Penalty_Goals_Per_90",
    "G+A-PK": "Non_Penalty_Goals_Plus_Assists_Per_90",
    "xG.1": "Expected_Goals_Per_90",
    "xAG.1": "Exp_Assisted_Goals_Per_90",
    "xG+xAG": "Exp_Goals_Plus_Exp_Assists_Per_90",     
    "npxG.1": "Non_Penalty_xG_Per_90",
    "npxG+xAG.1": "Non_Penalty_xG_Plus_Exp_Assists_Per_90",

}
dataset_standard = dataset_standard.rename(columns=rename_columns)

print(dataset_standard)
```

```python
# feature engineering – per-90 stats 
# add missing per-90 progressive stats
dataset_standard['Progressive_Carries_Per_90'] = dataset_standard['Progressive_Carries'] / dataset_standard['90s_Played']
dataset_standard['Progressive_Passes_Per_90'] = dataset_standard['Progressive_Passes'] / dataset_standard['90s_Played']
dataset_standard['Progressive_Passes_Received_Per_90'] = dataset_standard['Progressive_Passes_Received'] / dataset_standard['90s_Played']
dataset_standard['Yellow_Cards_Per_90'] = dataset_standard['Yellow_Cards'] / dataset_standard['90s_Played']
dataset_standard['Red_Cards_Per_90'] = dataset_standard['Red_Cards'] / dataset_standard['90s_Played']
dataset_standard['Penalties_Attempted_Per_90'] = dataset_standard['Penalties_Attempted'] / dataset_standard['90s_Played']
```

```python
print(dataset_standard)
print(dataset_standard.columns)
```

```python
#keep only per90 columns

#columns to merge with
cols_to_merge = ['Rank', 'Player', 'Nationality', 'Position', 'Age', 
           'Birth_Year', 'Matches_Played', 'Games_Started', 'Minutes','90s_Played']
#per_90 columns
per90_cols = [col for col in dataset_standard.columns if col.endswith('_90')]
#new dataset
dataset_standard_1 = dataset_standard[cols_to_merge + per90_cols]

print(dataset_standard_1)
print(dataset_standard_1.columns)
```

```python
#import defense stats table for the Top5 European 
#football for 2024/25 season
dataset_defense = pd.read_csv("fbref_big5_stats_24_25_def.csv")
print(dataset_defense)
```

```python
#rename columns in defense stats table to easy to undertand words using Fbref glossary 

rename_defensive_columns = {
    "Rk": "Rank",
    "Nation": "Nationality",
    "Pos": "Position",
    "Comp": "Competition",
    "Age": "Age",
    "Born": "Birth_Year",
    "90s": "90s_Played",

    #tackles
    "Tkl": "Tackles",             
    "TklW": "Tackles_Won",          
    "Def 3rd": "Tackles_Def_3rd", 
    "Mid 3rd": "Tackles_Mid_3rd", 
    "Att 3rd": "Tackles_Att_3rd", 

    #challenges 
    "Tkl.1": "Dribblers_Tackled", 
    "Att": "Dribbles_Challenged",
    "Tkl%": "%_Dribblers_Tackled",
    "Lost": "Challenges_Lost",      

    #blocks
    "Blocks": "Blocks",             
    "Sh": "Shots_Blocked",
    "Pass": "Passes_Blocked",

    # Interceptions, Clearances, Errors
    "Int": "Interceptions",
    "Tkl+Int": "Tackles_Plus_Interceptions",
    "Clr": "Clearances",
    "Err": "Errors_Leading_To_Shot"
}

dataset_defense = dataset_defense.rename(columns=rename_defensive_columns)

print(dataset_defense)
```

```python
#convert columns in the defense dataset to per90 
defensive_stats_per90 = [
    'Tackles',
    'Tackles_Won',
    'Tackles_Def_3rd',
    'Tackles_Mid_3rd',
    'Tackles_Att_3rd',
    'Dribblers_Tackled',
    'Dribbles_Challenged',
    'Challenges_Lost',
    'Blocks',
    'Shots_Blocked',
    'Passes_Blocked',
    'Interceptions',
    'Tackles_Plus_Interceptions',
    'Clearances',
    'Errors_Leading_To_Shot'
]

for col in defensive_stats_per90:
    new_col_name = f"{col}_Per_90"
    dataset_defense[new_col_name] = dataset_defense[col] / dataset_defense['90s_Played']
    
print(dataset_defense)
print(dataset_defense.columns)
```

```python
#keep only per90 columns

#columns to merge with
defs_cols_to_merge = ['Rank', 'Player', 'Nationality', 'Position', 'Age', 
           'Birth_Year','90s_Played']
#per_90 columns
per90_cols = [col for col in dataset_defense.columns if col.endswith('_90')]
#new dataset
dataset_defense_1 = dataset_defense[defs_cols_to_merge + per90_cols]

print(dataset_defense_1)
print(dataset_defense_1.columns)
```

```python
#merge both datasets and save

common_columns = ['Rank', 'Player', 'Nationality', 'Position', 'Age', 'Birth_Year', '90s_Played'
]

combined_stats_dataset = pd.merge(
    dataset_standard_1, 
    dataset_defense_1, 
    on=common_columns, 
    how='inner'
)

#save as csv file
combined_stats_dataset.to_csv('Combined_Stats_dataset.csv', index=False)

print(f"Final dataset shape: {combined_stats_dataset.shape}")
print(combined_stats_dataset)
```
